# Bar plot Of Toss winner top 3 tems

```python
plt.figure(figsize=(7,5))
plt.bar(list(ipl["toss_winner"].value_counts()[0:3].keys()),list(ipl["toss_winner"].value_counts()[0:3]),color=["b","m","y"])
plt.show()
```

# Make the pie plot of toss winning

```python
# pie plot 

plt.figure(figsize=(12,9))
plt.pie(list(ipl["toss_winner"].value_counts()),labels=list(ipl["toss_winner"].value_counts().keys()),autopct="%0.2f%%")
plt.show()
```
